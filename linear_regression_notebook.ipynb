```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
######## Step 1 ########
## Import the dataset
data_df = pd.read_csv('1.01. Simple linear regression.csv')
print(data_df.head())
print(data_df.describe())
```

```python
# Note: SAT sores include --> critical reading, math, and writing
# GPA --> is the grade point average at graduation
# Note: Why should we use SAT scores to predict GPA? Well, SAT is considered on eof the best predictors of intellectual capacity and capability.
# Note: It is safe to assume that this causual relationship is true.

y = data_df['GPA']
x = data_df['SAT']
```

```python
###### Step 2 ######
# Exploring Data
# It is important to explore the data before we start building the model. This allows you to understand the data and identify any potential issues. It also helps you to identify the relationship between the variables.
plt.scatter(x, y)
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
scatter_plot = plt.show()
scatter_plot
```

```python
###### Step 3 ######
# Regression itself
x = sm.add_constant(x)  # How we estimate the intercept b0
results = sm.OLS(y,x).fit() #creates a class object called results
print(results.summary())    #prints the summary of the regression and organizes into a table
```

```python
###### Step 4 ######
# Plotting the regression line

plt.scatter(x['SAT'], y)
y_hat = 0.0017*x['SAT'] + 0.275
fig = plt.plot(x['SAT'], y_hat, lw=4, c='blue', label='regression line')
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
regression_plot = plt.show()
regression_plot
```

```python
###### Step 5 ######
# Interpretation of the regression line - best fit
# We really want is the equation of the model, and a measure of its signifiane and explanation power
# y-intercept is 0.275
# coefficient is 0.0017 of SAT
print(results.params) #prints out the results of the regression model
```
